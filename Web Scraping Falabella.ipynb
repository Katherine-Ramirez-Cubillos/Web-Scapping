```python
#pagina donde va a buscar
#1. Falabella
driver.maximize_window()

docs = openpyxl.load_workbook ("Categorias.xlsx")
hoja = docs.get_sheet_by_name("Falabella")

itemsF = []
for columna in hoja.iter_rows():
    bien = columna[0].value
    itemsF.append(bien)
```

```python
from selenium.webdriver.common.keys import Keys

driver.get("https://www.falabella.com.co/falabella-co/search?Ntt=celular")

linksF = []
priceF = []
descF = []
codesF = []
timeF = []
obj = []
timeFF = []

for item in itemsF:
    searchBar = WebDriverWait(driver, 10).until(lambda x: x.find_element_by_class_name("jsx-1604206426.searchBar__Input")) #Selecionar la barra de busqueda
    searchBar.clear()
    searchBar.send_keys(item) #escribir en la barra de busqueda
    searchBar.send_keys(Keys.RETURN)
    print(item)
    time.sleep(5)
    n = "Nevera"
    l = "Lavadora"
    c = "Colchon"
    b = "Bicicleta"
    if item == n:
        elemsp = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".copy10.primary.high.jsx-185326735.normal"))   
    elif item == l:
        elemsp = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".copy10.primary.high.jsx-185326735.normal"))   
    elif item == c:
        elemsp = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".copy10.primary.high.jsx-185326735.normal"))   
    elif item == b:
        elemsp = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".copy10.primary.high.jsx-185326735.normal"))   
    elif item != n:
        elemsp = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".copy1.primary.high.jsx-185326735.normal"))
    #elif item != l:
        #elemsp = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".copy1.primary.high.jsx-185326735.normal"))
    for elemp in elemsp:
        if elemp.text != "":
            ele = elemp.text #obtiene el precio del webelement
            value = re.findall(r"[0-9.]*[0-9]+", ele) #quita el signo "$" del precio
            print(value[0])
            priceF.append(value[0]) #guarda el precio en una lista
        else:
            continue
    if item == n:
        elems = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".jsx-287641535.section-body.pod-link")) #busca y guarda los Webelemet con los precios en una lista
        for elem in itertools.islice(elems,None,None,1): #realiza un loop cada "2" Webelements
            linksF.append(elem.get_attribute("href")) #Obtiene y guarda el webelement en la lista "linksK"
            print(elem.get_attribute("href"))
            timeF.append(time.strftime("%d/%m/%y %H:%M:%S"))
            obj.append(item)
    elif item == l:
        elems = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".jsx-287641535.section-body.pod-link")) #busca y guarda los Webelemet con los precios en una lista
        for elem in itertools.islice(elems,None,None,1): #realiza un loop cada "2" Webelements
            linksF.append(elem.get_attribute("href")) #Obtiene y guarda el webelement en la lista "linksK"
            print(elem.get_attribute("href"))
            timeF.append(time.strftime("%d/%m/%y %H:%M:%S"))
            obj.append(item)
    elif item == c:
        elems = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".jsx-287641535.section-body.pod-link")) #busca y guarda los Webelemet con los precios en una lista
        for elem in itertools.islice(elems,None,None,1): #realiza un loop cada "2" Webelements
            linksF.append(elem.get_attribute("href")) #Obtiene y guarda el webelement en la lista "linksK"
            print(elem.get_attribute("href"))
            timeF.append(time.strftime("%d/%m/%y %H:%M:%S"))
            obj.append(item)
    elif item == b:
        elems = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".jsx-287641535.section-body.pod-link")) #busca y guarda los Webelemet con los precios en una lista
        for elem in itertools.islice(elems,None,None,1): #realiza un loop cada "2" Webelements
            linksF.append(elem.get_attribute("href")) #Obtiene y guarda el webelement en la lista "linksK"
            print(elem.get_attribute("href"))
            timeF.append(time.strftime("%d/%m/%y %H:%M:%S"))
            #timeFF.append(time.strftime("%d/%m/%y"))
            obj.append(item)
    elif item != n:
        elems = WebDriverWait(driver, 60).until(lambda x: x.find_elements_by_css_selector(".jsx-3773340100.jsx-3886284353.pod-link")) #busca y guarda los Webelemet con los precios en una lista
        for elem in itertools.islice(elems,None,None,2): #realiza un loop cada "2" Webelements
            linksF.append(elem.get_attribute("href")) #Obtiene y guarda el webelement en la lista "linksK"
            print(elem.get_attribute("href"))
            timeF.append(time.strftime("%d/%m/%y %H:%M:%S"))
            obj.append(item)
    print(len(priceF))
    print(len(linksF))
for link in linksF: #Obtener descripcion & Codigo unico de los links
    des = re.findall(r"[0-9]+.(.*?)/.", link)
    des = des[0]
    print(des)
    descF.append(des) #guarda cada descripcion en la lista "descK"
    code = re.findall(".*?([0-9]+)$",link)
    code = code[0]
    print(code)
    timeFF.append(time.strftime("%d/%m/%y"))
    codesF.append(code) #guarda cada descripcion en la lista "codesK"
```
